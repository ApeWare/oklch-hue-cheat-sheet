```python
# OKLCH Hue Wheel (merged "Red (0° & 360°)")
# Works in Google Colab, Jupyter, or locally with Python + matplotlib

import numpy as np
import matplotlib.pyplot as plt

def generate_hue_wheel(
    save_png="oklch_hue_wheel.png",
    save_svg="oklch_hue_wheel.svg",
    figsize=6,
    dpi=220
):
    # Create a polar plot
    fig, ax = plt.subplots(figsize=(figsize, figsize), subplot_kw={'projection': 'polar'})

    # 360 bars around the circle (one per degree)
    hues = np.linspace(0, 2*np.pi, 360, endpoint=False)
    # Using HSV colormap just to visualize hue positions; it’s a hue stand-in
    colors = plt.cm.hsv(np.linspace(0, 1, 360, endpoint=False))

    # Draw the ring
    ax.bar(hues, np.ones_like(hues), width=np.deg2rad(1), color=colors, edgecolor='none')

    # Primary spoke labels (merged red)
    label_map = {
        0:   "Red (0° & 360°)",
        60:  "Yellow (60°)",
        120: "Green (120°)",
        180: "Cyan (180°)",
        240: "Blue (240°)",
        300: "Magenta (300°)",
    }
    for angle_deg, text in label_map.items():
        ax.text(
            np.deg2rad(angle_deg), 1.08, text,
            ha='center', va='center',
            fontsize=10, fontweight='bold'
        )

    # Clean look
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_title("OKLCH Hue Cheat-Sheet", fontsize=14, fontweight="bold", pad=20)

    # Save
    plt.tight_layout()
    fig.savefig(save_png, dpi=dpi, bbox_inches="tight")
    fig.savefig(save_svg, dpi=dpi, bbox_inches="tight")
    plt.show()

# Generate files
generate_hue_wheel()

# If running in Google Colab, auto-prompt a download of the PNG:
try:
    from google.colab import files
    files.download("oklch_hue_wheel.png")
except Exception:
    pass  # Not in Colab; ignore
```
